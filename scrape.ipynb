```python
all_years_old
```

```python
def extract_pick(pick):
    if pd.notna(pick):
        match = re.search(r'\d+', pick)
        if match:
            return int(match.group())
    return pick
```

```python
# Split the 'Drafted (tm/rnd/yr)' column
drafted_info = all_years_old['Drafted (tm/rnd/yr)'].str.split(' / ', expand=True)

# Rename the columns
drafted_info.columns = ['Team', 'Round', 'Pick', 'Year']

# Concatenate the original DataFrame with the new 'drafted_info' DataFrame
all_years = pd.concat([all_years_old, drafted_info], axis=1)

# Drop the original 'Drafted (tm/rnd/yr)' column
all_years.drop(columns=['Drafted (tm/rnd/yr)', 'College'], inplace=True)

all_years.columns = all_years.columns.str.lower()
all_years['pick'] = all_years['pick'].apply(extract_pick)
all_years['pos'].value_counts()
```
